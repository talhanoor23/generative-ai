```python
corpus_2 = []
for i in range(len(sentences)):
    words = re.sub('[^a-zA-Z]',' ',sentences[i])
    words = words.lower()
    words = words.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    words = ' '.join(words)
    corpus_2.append(words)
```

# **Similarity_Recommendor_System**

```python
# Function to get top similar documents for a given document
def get_top_similar_docs(doc_index, top_n=3):
    similarity_scores = list(enumerate(similarity[doc_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Skip the first one (self-similarity)
    top_similar = similarity_scores[1:top_n+1]

    return [(f'Doc_{i}', score) for i, score in top_similar]
```

```python
similar_docs = get_top_similar_docs(0)
print(similar_docs)
```

```python
similar_docs = get_top_similar_docs(5)
print(similar_docs)
```

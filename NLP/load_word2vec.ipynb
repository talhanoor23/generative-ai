# **Load_dataset_model**

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import re
import string
import spacy
import nltk
import gensim
from bs4 import BeautifulSoup
```

```python
from gensim.models import Word2Vec
```

```python
df = pd.read_csv("/content/word2vec_df.csv")
```

```python
all_sentences = [sentence for sentences in df["processed_sentences"] for sentence in sentences]
```

```python
from gensim.models import Word2Vec

#easy wording and automatic. vocab_building and train occur automatically in one line of code.
Word2Vec_model = Word2Vec(sentences=all_sentences, window=10, min_count=2)
```

```python
def sentence_to_vector(sentences, model):
    """Converts sentences into a single vector by averaging word vectors."""
    vectors = [model.wv[word] for word in sentences if word in model.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(model.vector_size)

# 7️⃣ Apply Word2Vec Model to Each Review
df["review_vector"] = df["processed_sentences"].apply(lambda sentences: sentence_to_vector(sentences=sentences, model=Word2Vec_model))
```

# **Train_Test_Split**

```python
X = df['review_vector']
y = df['sentiment']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier()
# rf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf.fit(list(x_train), y_train)
y_pred = rf.predict(list(x_test))
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
